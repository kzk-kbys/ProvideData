`````python
import pandas as pd
import re
import datetime
`````

`````python
df_query_data = pd.read_csv("query_data.csv", index_col=None)
df_query_data.head()
`````

`````python
# トリガー開始
df_executing = df_query_data[df_query_data["message"].str.contains("Executing")]
func_executing = {}

for idx in range(len(df_executing)):
    timestamp = df_executing.iloc[idx]["timestamp [UTC]"]
    opeId = df_executing.iloc[idx]["operation_Id"]    
    func_executing[opeId] = pd.to_datetime(timestamp)
        
func_executing = sorted(func_executing.items(), key = lambda x : x[1], reverse=False)
operation_ids = [ opeId for opeId, _ in func_executing ]

func_execution_log = {}
for i, item in enumerate(func_executing):
    func_execution_log[item[0]] = {}
    func_execution_log[item[0]]["timestamp"] = item[1]
    func_execution_log[item[0]]["#Func"] = f"function{i+1}"
    
func_execution_log
`````

`````python
# トリガー終了
df_executed = df_query_data[df_query_data["message"].str.contains("Executed")]

for idx in range(len(df_executed)):
    msg = df_executed.iloc[idx]["message"]
    duration = re.match(".*?Duration=(\d+)ms", msg).group(1)
    opeId = df_executed.iloc[idx]["operation_Id"]  
        
    func_execution_log[opeId]["duration"] = duration
    
func_execution_log
`````

`````python
# 捌いたイベント数のカウント
df_events = df_query_data[df_query_data["message"].str.contains("messageBody")]

def count_events(opeId: str):
    return len(df_events[df_events["operation_Id"].str.contains(opeId)])
`````

`````python
# 同じ関数呼び出し単位（operation id）毎にログをまとめる
columns_func_log = ["#Func", "開始[msec]", "duration", "終了[msec]", "実行イベント数"]
df_func_log = pd.DataFrame(columns=columns_func_log)
df_func_log

for opeId in operation_ids:
    func_name = func_execution_log[opeId]["#Func"]
    timestamp = func_execution_log[opeId]["timestamp"]
    duration =  float(func_execution_log[opeId]["duration"])
    
    # 開始時刻msec
    # 最初のトリガーからの間隔とする（＝最初のトリガーはゼロ）
    trigger_duration = timestamp - func_execution_log[operation_ids[0]]["timestamp"]
    trigger_duration = trigger_duration.total_seconds() * 1000 # msec
    
    # 終了時刻msec
    # 開始時刻+duration
    trigger_end = trigger_duration + duration
    
    # 実行イベント数
    # 同じopeIdが付与されたイベントをカウントする
    # メッセージ内容の重複は一旦無視
    num_events = count_events(opeId)
        
    func_log = [func_name, trigger_duration, duration, trigger_end, num_events]   
    df_f = pd.DataFrame([func_log], columns=columns_func_log)
    df_func_log = pd.concat([df_func_log, df_f], ignore_index=True)
    
df_func_log
`````

`````python
# スループットを計算する
num_func = len(df_func_log)
total_dulation = float(df_func_log.loc[num_func-1, ["終了[msec]"]])

num_events = len(df_events)
throughput = 1000 * num_events / total_dulation

print(f"Throughput: {round(throuput)} [event/sec]")
`````
